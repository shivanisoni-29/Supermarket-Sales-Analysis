```python
import pandas as pd

df = pd.read_csv('sales.csv')

df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt

branch_sales = df.groupby('branch')['total_price'].sum()

plt.figure(figsize=(6,4))
branch_sales.plot(kind='bar')

plt.title('Total Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Sales')

plt.show()
```

```python
city_sales = df.groupby('city')['total_price'].sum()

plt.figure(figsize=(8,4))
city_sales.plot(kind='bar')

plt.title('Sales by City')
plt.xlabel('City')
plt.ylabel('Sales')

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,4))

sns.countplot(y='product_category', data=df)

plt.title('Product Category Distribution')

plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title('Customer Gender Distribution')
plt.show()
```

```python
top_products = df.groupby('product_name')['quantity'].sum()

top_products.sort_values(ascending=False).head(10).plot(kind='bar')

plt.title('Top 10 Selling Products')

plt.show()
```

```python
category_sales = df.groupby('product_category')['total_price'].sum()

category_sales.plot(kind='bar')

plt.title('Revenue by Product Category')

plt.show()
```
